```python
stores_df["Staff"].value_counts().plot(kind="bar")
```

```python
median = stores_df['Staff'].median()
```

```python
stores_df['Staff'] = stores_df['Staff'].replace(-2, median)
```

```python
stores_df['Staff'] = stores_df['Staff'].replace(300, median)
```

```python
stores_df['Staff'] = stores_df['Staff'].replace(600, median)
```

```python
stores_df["Staff"].value_counts().plot(kind="bar")
```

```python
plt.figure(figsize=(25,10))
cor =  stores_df.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt.show()
```

```python
# label_encoder object
var_lable = ["Car park", "Location", "Performance"]
label_encoder = LabelEncoder()
for col in stores_df.columns:
    if col in  var_lable:
        stores_df[col] = label_encoder.fit_transform(stores_df[col])
```

```python
X_features= stores_df.drop(["Performance","Town","Country","Store ID","Manager name", "Floor Space","30 min population","20 min population", "10 min population"], axis= 'columns')
y=stores_df['Performance']
```

```python
X_features.describe().T
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_features, y, test_size=0.2, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler().fit(X_train)
```

```python
X_train
```

```python
X_train_std = scaler.transform(X_train)
```

```python
X_test_std = scaler.transform(X_test)
```

```python
X_train_std_df = pd.DataFrame(X_train_std, index=X_train.index, columns=X_train.columns)
```

```python
X_test_std_df = pd.DataFrame(X_test_std, index=X_test.index, columns=X_test.columns)
```

```python
X_train_std_df.head()
```

```python
train_std_data = pd.concat([X_train_std_df, y_train], axis=1)
train_std_data.var().sort_values()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(train_std_data.corr(method='spearman'), annot=True)
```

```python
logistic_model = LogisticRegression()
```

```python
logistic_model.fit(X_train_std_df, y_train)
```

```python
y_predict= logistic_model.predict(X_test_std_df)
```

```python
print(y_predict)
```

```python
print("Accuracy: ", accuracy_score(y_test, y_predict))
```

```python
lr_score = logistic_model.score(X_test_std_df,y_test)
print(lr_score)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
cm = confusion_matrix (y_test, y_predict)

f, ax =plt.subplots(figsize=(5,5))
sns.heatmap(cm,annot=True,linewidths=0.5,linecolor="red",fmt=".0f",ax=ax)
plt.xlabel("predicted_y")
plt.ylabel("actual_y")
plt.show()
```

```python
print(classification_report(y_test, y_predict))
```
